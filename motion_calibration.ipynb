```python
import os
from motion_postprocess import extract_sample_frames, detect_motion
```

```python
BUFFER_DIR = "S:\\Dev\\rpi-surveillence\\buffer\\motion_test"
```

```python
segments = sorted(os.listdir(BUFFER_DIR)) #assumes a sortable order
print(len(segments))
m_total = 0
m_motion = 0
for i, seg in enumerate(segments):
    # only test new:
    # if seg.startswith("y_") or seg.startswith("n_") or seg.startswith("m_"):
    #     continue
    seg_path = os.path.join(BUFFER_DIR, seg)
    frames = extract_sample_frames(seg_path)
    motion = detect_motion(frames)
    if seg.startswith("y_"):
        assert motion
    if seg.startswith("n_"):
        assert not motion
    if seg.startswith("m_"):
        m_total += 1
        if motion:
            m_motion += 1
print(f"M motion clips: {m_motion}/{m_total}")    
```
